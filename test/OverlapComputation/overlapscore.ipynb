```python
from xbpy import rdutil
```

```python
mol1 = next(rdutil.read_molecules("test.pdb"))
mol2s = rdutil.read_molecules("test2.sdf")
```

```python
scores = []
for mol2 in mol2s:
    scores.append(rdutil.score_overlap(mol1, mol2))
```

```python
import pymolviz as pmv
```

```python
scores
```

```python
pmv.Expressions([f"test2 and state {i + 1}" for i in range(len(scores))], scores, transparencies=[0.5 for i in range(len(scores))]).write("color_by_similarity.py")
```
